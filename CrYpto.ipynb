```python
#import all files and libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

XLM = pd.read_excel(r'C:\Users\NoNoHome\Desktop\BGT-Crypto\XLM-20192020.xlsx')
ETH = pd.read_excel(r'C:\Users\NoNoHome\Desktop\BGT-Crypto\ETH-20192020.xlsx')
BTC = pd.read_excel(r'C:\Users\NoNoHome\Desktop\BGT-Crypto\BTC-20192020.xlsx')

#print(BTC)
```

```python
BTC_DayClose=BTC.loc["0":"739","Close"]
BTC_DayOpen=BTC.loc["0":"739","Open"]

XLM_DayClose=XLM.loc["0":"739","Close"]
XLM_DayOpen=XLM.loc["0":"739","Open"]

ETH_DayClose=ETH.loc["0":"739","Close"]
ETH_DayOpen=ETH.loc["0":"739","Open"]
```

```python
InvestDay1=500;
StartDayCash=InvestDay1
X= np.arange(1,741,1)

PercentDay = np.zeros((740,3))
# Start Compute 
PercentDay[:,0] =((BTC_DayClose-BTC_DayOpen)/BTC_DayOpen)*100
PercentDay[:,1] =((XLM_DayClose-XLM_DayOpen)/XLM_DayOpen)*100
PercentDay[:,2] =((ETH_DayClose-ETH_DayOpen)/ETH_DayOpen)*100
```

```python
Gain=np.zeros((3))
Loss=np.zeros((3))
print (Gain)
NoPainNoGain=np.zeros((3))
InvestDay1=100
StartDayCash=InvestDay1

CashInOut = np.zeros((740,3))
SumCash = np.zeros((740,3))
for i in range (0,3):
    
    count=0
    while (count < 739):
        count = count+1
        # Gain Day
        if PercentDay[count,i]>2.99:
            Gain[i]=Gain[i]+1
            CashInOut[count,i]=StartDayCash*PercentDay[count,i]/100
        
      # This has to be egal to previous day
            EndDayCash=StartDayCash
              # Loss Day
        elif PercentDay[count,i]<-9.99:
            Loss[i]=Loss[i]+1
            CashInOut[count,i]=StartDayCash*PercentDay[count,i]/100  
            EndDayCash=StartDayCash      
       
        else:
            NoPainNoGain[i]=NoPainNoGain[i]+1;   
       
        if count==0:
            SumCash[count,i]=SumCash[count,i]+CashInOut[count,i];

        else:
            SumCash[count,i]=SumCash[count-1,i]+CashInOut[count,i];

       
```

```python
print('invest Day1 per currency'); print(InvestDay1);print('GAin'); print(Gain);print('Loss'); print(Loss);print('NoPainNoGain'); print(NoPainNoGain);

# Creates two subplots and unpacks the output array immediately
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True)
ax1.stem(X,PercentDay[:,0])
ax1.set_title('BTC')
ax2.stem(X,PercentDay[:,1])
ax2.set_title('XLM')
ax3.stem(X,PercentDay[:,2])
ax3.set_title('ETH')

# Creates two subplots and unpacks the output array immediately
f1, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True)
ax1.stem(X,CashInOut[:,0])
ax2.stem(X,CashInOut[:,1])
ax3.stem(X,CashInOut[:,2])

# Creates two subplots and unpacks the output array immediately
f1, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True)
ax1.stem(X,SumCash[:,0])
ax2.stem(X,SumCash[:,1])
ax3.stem(X,SumCash[:,2])

print('Total Cashin');print(SumCash[739,:]) 
```
